```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import Data
df = pd.read_csv('kc_house_data.csv')
df.head(5).T
```

```python
#get some information about our Data-Set
df.info()
df.describe().transpose()
```

```python
# check if there are any Null values
df.isnull().sum()
```

```python
# drop some unnecessary columns
df = df.drop(['date','id','zipcode'],axis=1)
```

```python
X = df.drop('price',axis =1).values
y = df['price'].values
#splitting Train and Test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#standardization scaler - fit&transform on train, fit only on test
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train.astype(np.double))
X_test = s_scaler.transform(X_test.astype(np.double))
```

```python
print(X_train[:5])
```

```python
# Multiple Liner Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
#predicting the test set result
y_pred = regressor.predict(X_test)
mae_error = mean_absolute_error(y_test, y_pred)
print(mae_error)
```

```python
#compare actual output values with predicted values
y_pred = regressor.predict(X_test)
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df1.head(10))

# evaluate the performance of the algorithm (MAE - MSE - RMSE)
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
kNN = KNeighborsRegressor(n_neighbors=5)

kNN.fit(X_train, y_train)
y_pred = kNN.predict(X_test)
mae_error = mean_absolute_error(y_test, y_pred)
print(mae_error)
```

```python
# Creating a Neural Network Model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
#model.add(Dense(16,activation='relu'))
model.add(Dense(1)) #If you don't specify anything, no activation is applied (ie. "linear" activation: a(x) = x).
model.compile(optimizer='Adam',loss=tf.keras.losses.MeanAbsoluteError())
```

```python
model.fit(x=X_train,y=y_train, validation_split=0.1,
          batch_size=128,epochs=150)
```

```python
model.summary()
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot(figsize=(12,8))
```

```python
#compare actual output values with predicted values
y_pred = model.predict(X_test)

# evaluate the performance of the algorithm (MAE - MSE - RMSE)
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```
